```python
import pandas as pd

df = pd.read_excel('planilha_tratada_oficial.xlsx')
df
```

```python
df['Subprefeitura'] = df['Subprefeitura'].str.replace('ARICANDUVA/FORMOSA', 'ARICANDUVA/FORMOSA/CARRÃO').str.replace('ARICANDUVA / VILA FORMOSA', 'ARICANDUVA/FORMOSA/CARRÃO').str.replace("M'BOI IRIM", "M'BOI MIRIM").str.replace('M BOI MIRIM ', "M'BOI MIRIM")
```

```python
df['Subprefeitura'] = df['Subprefeitura'].str.replace('PARELHEIROS ', 'PARELHEIROS').str.replace('BUTANTA', 'BUTANTÃ').str.replace('PIRITUBA-JARAGUA', 'PIRITUBA/JARAGUA')
```

```python
df['Subprefeitura'] = df['Subprefeitura'].str.replace('ARICANDUVA/FORMOSA/CARRÃO/CARRÃO', 'ARICANDUVA/FORMOSA/CARRÃO')
```

```python
df['Subprefeitura'] = df['Subprefeitura'].str.replace('CAPELA DO SOCORRO ', 'CAPELA DO SOCORRO')
```

```python
df['Subprefeitura'] = df['Subprefeitura'].str.replace('SANTANA / TUCURUVI', 'SANTANA/TUCURUVI').str.replace('V. MARIA / V. GUILHERME', 'VILA MARIA/VILA GUILHERME').str.replace('JAÇANÃ / TREMEMBE', 'JAÇANÃ/TREMEMBE')
```

```python
df['Subprefeitura'] = df['Subprefeitura'].str.replace('PIRITUBA/JARAGUA/JARAGUA', 'PIRITUBA/JARAGUA')
```

```python
df['Subprefeitura'].unique()
```

```python
subprefeituras_por_zona = {
    'Zona Sul': [
        'CAMPO LIMPO',
        'CAPELA DO SOCORRO',
        "M'BOI MIRIM",
        'CIDADE ADEMAR',
        'PARELHEIROS',
        'SANTO AMARO',
        'IPIRANGA',
        'JABAQUARA'
    ],
    'Zona Norte': [
        'PERUS',
        'PIRITUBA/JARAGUA',
        'CASA VERDE',
        'FREGUESIA DO Ó',
        'SANTANA/TUCURUVI',
        'JAÇANÃ/TREMEMBE',
        'VILA MARIA/VILA GUILHERME'
    ],
    'Zona Leste': [
        'ARICANDUVA/FORMOSA/CARRÃO',
        'CIDADE TIRADENTES',
        'ERMELINO MATARAZZO',
        'GUAIANASES',
        'ITAQUERA',
        'ITAIM PAULISTA',
        'PENHA',
        'SÃO MIGUEL PAULISTA',
        'SAPOPEMBA',
        'SÃO MATEUS',
        'VILA PRUDENTE'
    ],
    'Zona Oeste': [
        'LAPA',
        'BUTANTÃ',
        'PINHEIROS'
    ],
    'Zona Centro': [
        'SÉ',
        'MOOCA',
        'VILA MARIANA'
    ]
}

def zonas(subprefeitura):
    for zona, subpre in subprefeituras_por_zona.items():
        for subp in subpre:
            if subprefeitura == subp:
                return zona

        
df['Zona'] = df['Subprefeitura'].apply(zonas)
```

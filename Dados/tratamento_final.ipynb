```python
import pandas as pd

df = pd.read_excel('Planilhas_Unificadas/planilha_unificada_volumetria_ano.xlsx')
df
```

```python
df['Quadrimestre'] = df['Origem_aba'].str.replace('º Quadrimestre', '')
df['Quadrimestre'] = df['Quadrimestre'].apply(int)
df['Quadrimestre']
```

```python
df = df.drop(columns='Origem_aba')
df = df[['Endereço', 'Sigla_Sub', 'Subprefeitura', 'Lote', 'Contratada', 'Volume', 'Lat', 'Long', 'Volume_int', 'Quadrimestre', 'Ano']]
df
```

```python
df.to_excel('planilha_tratada.xlsx', index=0)
```

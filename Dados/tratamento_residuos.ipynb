```python
import pandas as pd

df = pd.read_excel('residuos_final.xlsx', sheet_name='2019')
df1 = pd.read_excel('residuos_final.xlsx', sheet_name='2020')
df2 = pd.read_excel('residuos_final.xlsx', sheet_name='2021')
df2
```

```python
df2
```

```python
df.drop(index=28, inplace=True)
df.drop(['Total 2019'], axis=1, inplace=True)

df1.drop(index=32, inplace=True)
df1.drop(['Total 2020'], axis=1, inplace=True)

df2.drop(index=32, inplace=True)
```

```python
dict = {
    'tipo': [],
    'quantidade': [],
    'mes': [],
    'ano': []
}
```

```python
meses = ['Janeiro',	'Fevereiro','Março','Abril'	,'Maio'	,'Junho','Julho'	,'Agosto',	'Setembro',	'Outubro'	,'Novembro',	'Dezembro']

index = 0

for mes in meses:
    for row in df[mes]:
        dict['tipo'].append(df['Tipo de resíduo - Toneladas'][index])
        dict['quantidade'].append(row)
        dict['ano'].append(2019)
        dict['mes'].append(mes)
        index += 1
    index = 0
```

```python
meses = ['Janeiro',	'Fevereiro','Março','Abril'	,'Maio'	,'Junho','Julho'	,'Agosto',	'Setembro',	'Outubro'	,'Novembro',	'Dezembro']

index = 0

for mes in meses:
    for row in df1[mes]:
        dict['tipo'].append(df1['Tipo de resíduo - Toneladas'][index])
        dict['quantidade'].append(row)
        dict['ano'].append(2020)
        dict['mes'].append(mes)
        index += 1
    index = 0

                
```

```python
meses = ['Janeiro',	'Fevereiro']

index = 0

for mes in meses:
    for row in df2[mes]:
        dict['tipo'].append(df2['Tipo de resíduo - Toneladas'][index])
        dict['quantidade'].append(row)
        dict['ano'].append(2021)
        dict['mes'].append(mes)
        index += 1
    index = 0
```

```python
df_unificado = pd.DataFrame(dict)
df_unificado
```

```python
df_unificado.to_excel('residuos_unificado.xlsx', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
df=pd.read_excel('planilha_unificada.xlsx')
df
```

```python
df.drop(['Fonte','Data_ref'],axis= 1,inplace=True)
df
```

```python


def tratar_ano(anos):
    texto = str(anos).strip().upper()

    if '2019' in texto:
        return 1
    elif '2021' in texto:
        return 5
    elif '2020' in texto:
        return 3
    elif 'REVITALIZADO' in texto or 'ELIMINADO' in texto:
        return None

df['Ano'] = df['Origem_aba'].apply(tratar_ano).astype('Int64')

df.to_excel('planilha_unificada_volumetria_tratado.xlsx', index= False)
```
